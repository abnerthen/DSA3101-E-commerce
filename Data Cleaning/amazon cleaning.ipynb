# **2. Drop redundant columns**

```python
df = df.drop(['product_link', 'img_link', 'review_content', 'review_title', 'about_product'], axis = 1)
df.info()
#removed 5 redundant columns for analysis
```

# **3. Remove abnormal datapoints**
- Drop rows where rating_count is NA
- Drop rows where rating is '|'

```python
df.isnull().sum()
# We note that there are 2 rows with no rating_count but has a rating
```

```python
df.dropna(subset=['rating_count'], inplace = True)
#Removed 2 rows with no rating_count
```

```python
# We notice a row where rating = |  which is invalid, so we remove it

df = df[~df['rating'].str.contains('\|', na=False)]
```

# **4. Drop duplicate rows**

```python
# Drop duplicate rows
df = df.drop_duplicates()
```

# **5. Remove currency symbol**

```python
check1 = df['discounted_price'].str[0].value_counts()
check1
# check if currency is the same
```

```python
check2 = df['actual_price'].str[0].value_counts()
check2
# check if currency is the same
```

```python
df['discounted_price_modified'] = df['discounted_price'].str[1:]
df['actual_price_modified'] = df['actual_price'].str[1:]

#price is in rupees
#remove currency symbol from original discount price and actual price column since currency is the same

#remove original price columns with currency symbol
df = df.drop(['actual_price', 'discounted_price'], axis = 1)
```

# **6. Convert numerical columns to float datatype**

Double check datatypes of other columns

```python
#remove comma in price and change from object to float,
#change rating to float

import re

def parse_to_float(_str):
  return float(re.sub(",", "", _str))

def parse_to_int(_str):
  return int(re.sub(",", "", _str))

df['discounted_price_modified'] = df['discounted_price_modified'].apply(parse_to_float)
df['actual_price_modified'] = df['actual_price_modified'].apply(parse_to_float)
df['rating'] = df['rating'].apply(parse_to_float)

#convert rating_count to int and removing the commas in the rating_count column

df['rating_count'] = df['rating_count'].str.replace(',', '')
df['rating_count'] = pd.to_numeric(df['rating_count'])
df['rating_count'] = df['rating_count'].fillna(0).astype(int)

#converting discount percentage column into an integer

df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'])
df['discount_percentage'] = df['discount_percentage']/100

df[['discounted_price_modified', 'actual_price_modified', 'rating', 'rating_count', 'discount_percentage']]
```

```python
#function to check if a column type is a string, since pandas displays everything as an object
def is_string_column(df, column_name):
    return df[column_name].apply(lambda x: isinstance(x, str)).all()
```

# **7. Split 'category' columns into sub categories**

```python
#splitting into further categories

df_split = df['category'].str.split('|', expand=True)
df_split.columns = [f'category_{i+1}' for i in range(df_split.shape[1])]
df = pd.concat([df, df_split], axis=1)
df.drop('category', axis=1, inplace=True)
df.head()
```
